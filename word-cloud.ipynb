```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np

# Sample text
text = """
Leader
Leader
Prioritizer
Strategic
Connector
Adaptable
Communicator
Innovator
Catalyst
Catalyst

SQL
Python
Rtidyverse
SAS
Jupyter
Tableau
PowerBI
Office

MarketingAnalytics
MarketingAnalytics
AdobeAnalytics
GoogleAnalytics
CustomerAnalytics
CustomerAnalytics
ActionIQ
CDP
Dynamics
AWS
AWS
GCP
Azure
Cloud

ABtesting
MachineLearning
MachineLearning
MachineLearning
DeepLearning
Statistics
LLMs
"""

# Create a WordCloud object
wordcloud = WordCloud(
    width=5600,
    height=1400,
    background_color='#1D2226',
    colormap='Pastel1',
    max_words=100
).generate(text)

# Display the generated image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.tight_layout()
plt.show()
```
